Importanto as libs necessárias

```python
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
```

carregando o dataset já separado para treino e testes

```python
(x_train, y_train),(x_test, y_test) = datasets.mnist.load_data()
# Normalizar os valores dos pixels para o intervalo 0-1
x_train = x_train / 255.0
x_test = x_test / 255.0
```

Criando uma CNN

```python
model = models.Sequential([
    # layer de convolução 32 filtros
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    # layer de pooling 2x2
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    # layer totalmente conectada (para padrões globais)
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Treinando o modelo

```python
model.fit(x_train, y_train, epochs=3)
```

Avaliando a acurácia

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f'Acuracia: {test_acc}')
```

Outros testes

```python
for i in range(100, 105):
  img_array = np.expand_dims(x_test[i], axis=0)
  predictions = model.predict(img_array, verbose=0)
  print(np.argmax(predictions, axis=1))
  plt.imshow(x_test[i])
  plt.show()
  print("\n----------\n")
```
